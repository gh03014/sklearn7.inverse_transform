```python
#fas-정재환
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np
import csv
import json
import seaborn as sns
%matplotlib inline 
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("NHIS_OPEN_GJ_2017_3.csv", encoding='cp949', nrows=50000)
df
```

```python
# StandardScaler()를 사용해 평균0, 분산1로 표준화 처리
x_원본 = df[['체중']]
scaler = StandardScaler()
x_표준화 = scaler.fit_transform(x_원본)
x_표준화
```

```python
# 표준화된 데이터의 최대값과 최소값 확인
print('표준화 데이터 최대값: ' + str(x_표준화.max()))
print('표준화 데이터 최소값: ' + str(x_표준화.min()))
print('표준화 데이터 평균값: ' + str(x_표준화.mean()))
```

```python
# inverse_transform()을 사용해 표준화된 데이터를 다시 원래대로 복원
x_복원 = scaler.inverse_transform(x_표준화)
x_복원
```

```python
# 실제로 제대로 복원이 되었는지 원본값과 최댓값, 최솟값, 평균값을 비교해보자
x_원본 = x_원본['체중']
print('원본 데이터 최소값: ' + str(x_원본.min()))
print('복원 데이터 최소값: ' + str(x_복원.min()))
print('원본 데이터 최대값: ' + str(x_원본.max()))
print('복원 데이터 최대값: ' + str(x_복원.max()))
print('원본 데이터 평균값: ' + str(x_원본.mean()))
print('복원 데이터 평균값: ' + str(x_복원.mean()))
```
